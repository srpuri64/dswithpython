## Data visualizations

When we want to visualize our data, it can be extremely tricky to identify the right visualization for our dataset. One approach is to choose the visualization based on the function you want to perform: 

-  **Comparison**: Compare variables
-  **Relationship**: Understand relationships or how variables relate to each other
-  **Distribution**: Understand how the values within a variable or a dataset are distributed
-  **Trend Over Time**: Identify a pattern or trend over time

### Relationship Plots
Let's create visualizations to better understand the relationship of the features in the dataset.

#### Pair Plot
This plot is a type of visualization that shows scatterplots of all possible feature combinations. Also, colors the points based on the target class (Iris species). It helps us visualize how features relate to each other. Visually, it is a matrix of scatterplots, showing the relationships between all pairs of numeric variables in a dataset. It is an excellent tool for visualizing multivariate data!

```python
# Create a new data frame where the column ID is missing
iris_df = iris.drop('Id', axis=1)

# Create a pair plot
pair_plot = sns.pairplot(iris_df, hue='Species', markers='+')

# Add a title to the pair plot
plt.suptitle("Pair Plot of Iris Data by Species", y=1.02)

plt.show()
```

In this example, hue="species" colors the data points by the species of the iris flowers, making it easy to distinguish different classes.

We learn that: 

- **PetalLength** and **PetalWidth** are significantly more useful attributes to identify various flower types. We can see, that the Iris-setosa species is significantly different in terms of petal length and width from its other two counterparts. If the Petal Width of an Iris flower is less than or equal to approximately 0.5 cms, we can classify it as an Iris-setosa.
- **SepalLength** and **SepalWidth** are quite overlapping for all three species. They may not be the best features to use for modeling purposes due to their similarity.
- The relationship between pairs of features of Iris-Setosa (blue) is distinctly different from those of the other two species. 
- We observe some overlap in the pairwise relationships of the other two species, Iris-versicolor (orange) and Iris-virginica (green).

### PairGrid
A PairGrid is a versatile tool for creating custom pair plots. It allows you to control the type of plots in the diagonal and off-diagonal cells. With PairGrid, you can customize the type of plots displayed on the diagonal and off-diagonal cells.

```python
# Create a PairGrid
g = sns.PairGrid(iris_df, hue="Species")

#Customize a PairGrid
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)

# Add a legend and a title
g.add_legend()
plt.suptitle("Customized PairGrid of Iris Dataset", y=1.02)

plt.show()
```

PairGrids are especially useful when you have multiple variables and you want to visualize how they relate to each other. Here's how to interpret the components of a PairGrid:

- **Main diagonal (diagonal plots)**: The main diagonal of the PairGrid typically contains histograms or other univariate plots for each variable. These plots show the distribution of each variable in isolation. You can interpret these plots to understand the individual variable's distribution and characteristics.
- **Off-diagonal plots (Non-diagonal cells)**: The off-diagonal cells of the PairGrid display scatterplots, regression plots, or other bivariate plots showing the relationship between pairs of variables. Each off-diagonal cell compares one variable to another. These plots help you understand how variables interact with each other. Interpretations of off-diagonal plots include:
- **Scatterplots**: Scatterplots in the off-diagonal cells show how two variables are related. You can look for patterns, trends, and relationships between the variables. For example, you might assess if there's a linear or nonlinear relationship between two variables.
- **Regression plots**: Some PairGrids include regression lines or fits on the scatterplots, which allow you to assess the strength and direction of the relationship between the variables.
- **Upper triangle and lower triangle**: In symmetric PairGrids, you'll often see mirrored plots across the main diagonal. The plots in the upper triangle are reflections of those in the lower triangle. They convey the same information, so you can focus on either the upper or lower triangle.
- **Customization and styling**: PairGrids can be customized to include additional information, such as color-coding by a categorical variable (e.g., hue), making it easier to interpret how the relationships differ for different groups.
- **Legend and titles**: Interpret any legends, titles, or axis labels provided in the PairGrid. Legends often explain how variables are color-coded or grouped, while titles and labels provide context for what's being plotted.
- **Patterns and outliers**: Look for patterns or outliers in the off-diagonal scatterplots that might suggest interesting relationships or data anomalies.
- **Symmetry and correlations**: In some cases, you can assess the symmetry of off-diagonal plots. For example, if you have the same variable on both the x and y axes of a scatterplot, you'll see a diagonal line. Deviations from this line indicate correlation between a variable and itself.
- **Annotations and trends**: Pay attention to any annotations or trend lines that provide additional information about the relationships between variables.

An example of customizing PairGrid with scatterplots, and KDE:

```python
# Create a PairGrid for the 'iris' dataset
g = sns.PairGrid(iris, diag_sharey=False)

# Map a scatterplot to the upper triangle of the PairGrid
g.map_upper(sns.scatterplot, s=15)

# Map a kernel density estimate (KDE) plot to the lower triangle of the PairGrid
g.map_lower(sns.kdeplot)

# Map a kernel density estimate (KDE) plot to the diagonal of the PairGrid
g.map_diag(sns.kdeplot, lw=2)
```

#### Heatmap
A heatmap is used to visualize the correlation between variables in a dataset. Essentially it tells us if two or more variables are related and not and also reveals the strength of their relationship through colors.

```python
# Compute the correlation matrix for numeric columns
numeric_columns = iris.select_dtypes(include=['float64'])  # Select only numeric columns
correlation_matrix = numeric_columns.corr()

# Create a heatmap
sns.heatmap(correlation_matrix, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

From the plot, we learn that: 

- The **SepalWidth** and **SepalLength** are not correlated.
- The **PetalLength** and **PetalWidth** are positively correlated and the strength of the color reveals that their correlation is strong.
- The **PetalWidth** and **SepalLength** have a positive correlation.
- **Sepalwidth** and **PetalLength** are negatively correlated. 

### Distribution Plots
Distribution plots are used to represent the distribution or frequency of values within a single variable or dataset. It shows how data is spread. Common types of distribution plots include: 
- Histograms
- Boxplots
- Kernel Density Estimation (KDE) plots

which provide insights into the shape, central tendency, spread, and other characteristics of the data's distribution.

#### Box plot

```python
# Create a boxplot for the Iris dataset
plt.figure(figsize=(10, 6))
boxplot = sns.boxplot(data=iris_df, orient="h")

plt.title("Boxplot of Iris Dataset")
plt.show()
```

To interpret a boxplot, we pay attention to:  
- **Box and Whiskers Structure**: The boxplot consists of a rectangular box and two whiskers extending from it.
- **Box Center**: The box represents the interquartile range (IQR) and contains the median (50th percentile) of the data. It gives insights into the central tendency of the dataset.
- **Whiskers**: The whiskers extend from the box to the minimum and maximum values within a specified range, excluding outliers.
- **Outliers**: Individual data points beyond the whiskers are considered outliers and are displayed as separate points. Outliers can be indicative of unusual or extreme data values.
- **IQR and Spread**: The length of the box (IQR) indicates the spread of the central 50% of the data.
- **Symmetry and Skewness**: The symmetry or skewness of the data can be inferred from the box's position within the plot. A centered box suggests symmetry, while an offset box suggests skewness.

The boxplot provides us with:
- Minimum score: The minimum value in the dataset.
- First quartile: Information on the 25% of the dataset.
- Median: Information on the 50% of the dataset.
- Third Quartile: Information on 75% of the dataset.
- Maximum Score: The highest value in the dataset.

From the boxplots, we can see that some features (e.g., petal length and petal width) have significant differences in their distributions among the three species, making them valuable for classification.

#### Histogram

```python
# Histograms for each species
sns.histplot(data=iris, x='PetalLengthCm', hue='Species')

# Add title
plt.title("Histogram of Petal Lengths, by Species")
```

How you can interpret a histogram:
- **Data Distribution**: A histogram allows you to quickly grasp the overall distribution of the data. You can see whether the data is skewed to the left (negative skew), skewed to the right (positive skew), or relatively symmetrical (bell-shaped).
- **Central Tendency**: The center of the histogram typically corresponds to the central tendency of the data. For example, in a normal distribution, the peak of the histogram represents the mean (average) of the dataset.
- **Spread/Dispersion**: The width of the histogram bars reflects the spread or dispersion of the data. A wide histogram suggests greater variability, while a narrow one indicates less variability.
- **Outliers**: Outliers or extreme values are often visible as bars that are significantly taller or shorter than the rest. They stand out as unusual data points.
- **Bimodal or Multimodal Distributions**: If you see multiple peaks (high bars) in the histogram, it may indicate that the data has more than one mode, meaning it is bimodal or multimodal.
- **Skewness**: The direction of skewness can be determined by the longer tail of the histogram. If the tail is on the right, the distribution is positively skewed, and if it's on the left, it's negatively skewed.
- **Gaps and Gaps Between Bars**: Gaps between bars in the histogram can indicate discontinuities in the data. It can be useful for identifying missing data or errors.
- **Range**: The range of the x-axis (horizontal) represents the entire range of values in the dataset.

#### Density plots
The next type of plot is a kernel density estimate (KDE) plot. In case you're not familiar with KDE plots, KDE is used to estimate the probability density function of a dataset. In simpler terms, it helps you understand how likely different values are in your data. It provides a smooth curve that represents the data distribution. KDE plots to compare the distribution of two or more groups to see if they are significantly different from each other.

```python
# KDE plots for each species
sns.kdeplot(data=iris, x='PetalLengthCm', hue='Species', fill=True)

# Add title
plt.title("Distribution of Petal Lengths, by Species")
```

An interesting pattern to be observed in the plots is that the Iris species seem to belong to one of two groups instead of three, where Iris versicolor and Iris virginica seem to have overlapping values for Petal Length, while Iris setosa belongs in a category all by itself.

According to the values of Petal Length, we might even be able to classify an Iris flower as Iris setosa (as opposed to its counterparts Iris versicolor or Iris virginica) just by looking at the Petal Length. If the Petal Length of an iris flower is less than or equal to 2 cm, it's most likely to be an Iris setosa specie!

```python
# 2D KDE plot
sns.jointplot(x=iris['PetalLengthCm'], y=iris['SepalWidthCm'], kind="kde")
```

A two-dimensional KDE plot eliminates the restriction of visualizing a single column when creating a KDE plot. You can create a 2D KDE plot using the **sns.jointplot** command. In this plot, colors indicate the likelihood of different combinations of Sepal Width and Petal Length. Darker areas suggest a higher likelihood.

#### Violin Plot
A violin plot combines a box plot with a KDE (Kernel Density Estimate) plot. Here, we're comparing the sepal length of different Iris species using violin plots. KDE is a way to show the frequency of data just like a histogram. A limitation of the histogram is that it depends on the bins and their sizes. If there is a change in size/bins, consequently, the shape changes too. For that reason, KDE is considered a good estimator of the distribution.

```python
# Create a violin plot
sns.violinplot(x="Species", y="SepalLengthCm", data=iris, inner="quart")
plt.title("Violin Plot of Sepal Length by Species")
plt.show()
```

Violin plots help us:

- Understand the distribution of the data. It enables us to visualize the distribution of a numeric variable for one or several groups.
- Estimate the density of the feature. The width of each plot provides insight into the density of the feature, which means that the violin plot is used to provide an estimate of the probability density function (PDF) for a specific feature or variable in the dataset. The PDF represents how likely different values of that variable are to occur.
- Compare multiple groups. It can be a tool for comparing the distribution of a variable across multiple groups.

Here's how to interpret it:

- **Width of the violin**: The width of the violin at any given point along the x-axis represents the estimated density or probability of the variable at that value. A wider section of the violin indicates a higher density of data points, while a narrower section indicates a lower density.
- **Symmetry and shape**: The shape of the violin plot can be symmetrical or asymmetrical, depending on the distribution of the data. A symmetrical violin plot suggests that the data is relatively evenly distributed around the central value, while an asymmetrical shape might indicate skewness in the data.
- **Central line**: The central line within the violin plot typically represents the median or mean of the data, providing a measure of central tendency.
- **Interquartile range**: The thicker portion of the violin, often referred to as the "body," represents the interquartile range (IQR) of the data, giving insights into the spread of data around the median.

#### Swarm Plot
A swarm plot is a categorical scatter plot that displays individual data points along a category axis. It works for continuous and categorical variables. It is a type of scatter plot but the scatter plots visualize overlapping values which is eliminated here. It gives insights into the distribution of the data. The drawback is that they work for a small set of data. In large sets, getting insights from swarm plots can be difficult. Also, they don't show statistics (like median or quartiles) like boxplots and violin plots.

```python
# Create a swarm plot
sns.swarmplot(x="Species", y="PetalLengthCm", data=iris)
plt.title("Swarm Plot of Petal Length by Species")
plt.show()
```

### Common Issues and Challenges

While Seaborn is a versatile and user-friendly data visualization library, users can encounter common issues and challenges while working with it:
- **Compatibility with Matplotlib**:
    - Conflicting Customizations: Be careful when customizing both Seaborn and Matplotlib in the same plot.
- **Data cleaning and formatting**:
    - Data preparation: Seaborn expects data in a specific format, and you may need to preprocess your data to handle missing values.
- **Plotting large datasets**:
    - Performance: Seaborn may be slower with large datasets. Consider subsampling or aggregating data.
- **Customization challenges**:
    - Customization Limitations: In some cases, you might need to use Matplotlib directly for fine-grained control.
- **Understanding plot types**:
    - Choosing the right plot: Understand the characteristics of your data to choose the right visualization technique.
- **Interpreting statistical plots**:
    - Understanding Statistics: Some Seaborn plots display statistical summaries of data, and you may need to understand these statistics to interpret the plots correctly.
